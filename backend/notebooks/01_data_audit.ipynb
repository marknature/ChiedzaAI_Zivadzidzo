# ZivaDzidzo data audit
Run this before training. It records schema, missingness, class balance, and the production privacy boundary.

```python
import pandas as pd
from ml_config import INDUSTRY4_FILE, READINESS_TARGET, GAP_TARGET
df = pd.read_csv(INDUSTRY4_FILE)
print(df.shape)
display(df.head())
```

```python
audit = pd.DataFrame({'dtype': df.dtypes.astype(str), 'missing': df.isna().sum(), 'unique': df.nunique()})
display(audit.sort_values('missing', ascending=False))
display(df[READINESS_TARGET].value_counts(normalize=True).rename('proportion'))
display(df[GAP_TARGET].describe())
```
